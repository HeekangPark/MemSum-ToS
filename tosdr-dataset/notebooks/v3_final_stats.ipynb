```python
import json
import pandas as pd
import pickle
from tqdm import tqdm


def dump_pickle(file_path, object):
    with open(file_path, 'wb') as handle:
        pickle.dump(object, handle)
```

```python
stats = []

with open("train_labelled.jsonl", 'r') as f:
    for line in tqdm(f):
        line = json.loads(line)
        text = line["text"]
        summary = line["summary"]
        oracles = line["indices"]
        scores = line["score"][1:]
        num_sent_text = len(text)
        num_sent_summary = len(summary)
        num_oracles = len(oracles)

        
        avg_word_text, avg_word_summary, avg_score = .0, .0, .0
        for sent in text:
            words_per_sent = sent.split()
            avg_word_text += len(words_per_sent)
        for sent in summary:
            words_per_sent = sent.split()
            avg_word_summary += len(words_per_sent)
        for score in scores:
            avg_score += score
        avg_word_text /= num_sent_text
        avg_word_summary /= num_sent_summary
        avg_score /= num_oracles
        
        stats.append({"num_sent_text": num_sent_text,
                      "num_sent_summary": num_sent_summary,
                      "num_oracles": num_oracles,
                      "avg_word_text": avg_word_text,
                      "avg_word_summary": avg_word_summary,
                      "avg_score": avg_score})

train_df = pd.DataFrame.from_records(stats)
dump_pickle("./train_labelled_stats.pickle", train_df)
```

```python
stats = []

with open("validation.jsonl", 'r') as f:
    for line in tqdm(f):
        line = json.loads(line)
        text = line["text"]
        summary = line["summary"]
        num_sent_text = len(text)
        num_sent_summary = len(summary)
        
        avg_word_text, avg_word_summary = .0, .0
        for sent in text:
            words_per_sent = sent.split()
            avg_word_text += len(words_per_sent)
        for sent in summary:
            words_per_sent = sent.split()
            avg_word_summary += len(words_per_sent)
        avg_word_text /= num_sent_text
        avg_word_summary /= num_sent_summary
        
        stats.append({"num_sent_text": num_sent_text,
                      "num_sent_summary": num_sent_summary,
                      "avg_word_text": avg_word_text,
                      "avg_word_summary": avg_word_summary})

validation_df = pd.DataFrame.from_records(stats)
dump_pickle("./validation_stats.pickle", validation_df)
```
